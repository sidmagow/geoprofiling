```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
from sklearn.cluster import KMeans as km
import sklearn.utils as skut
import gmplot as gm
import geocoder
```

```python
data=pd.read_csv("Final.csv")
```

```python
data
```

```python
lis=['District', 'Crime', 'Timestamp']
for c in lis:
    data.drop(c, axis=1, inplace=True)
```

```python
data
```

```python
def remove_outliers():
    for i in range(len(data.index)):
        if data['Latitude'][i] < 28.412782:
            data['Latitude'][i] = 28.412782
        elif data['Latitude'][i] > 28.88:
            data['Latitude'][i] = 28.88
        if data['Longitude'][i] < 76.838084:
            data['Longitude'][i] = 76.838084
        elif data['Longitude'][i] > 77.323460:
            data['Longitude'][i] = 77.323460
```

```python
remove_outliers()
```

```python
data['Latitude'].size
```

```python
data=skut.shuffle(data)
```

```python
data
```

```python
data['Latitude'].mean()
```

```python
data['Longitude'].mean()
```

```python
kmeans = km(n_clusters=18).fit(data)
```

```python
prediction=kmeans.predict(data)
```

```python
centers=kmeans.cluster_centers_
```
