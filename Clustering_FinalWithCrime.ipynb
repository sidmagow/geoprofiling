```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
from sklearn.cluster import KMeans as km
import sklearn.utils as skut
import gmplot as gm
import geocoder
```

```python
data=pd.read_csv("FinalGeoDataSetWithCrime.csv")
```

```python
data
```

```python
lis=['District']
for c in lis:
    data.drop(c, axis=1, inplace=True)
```

```python
def f(s):
    s1=s[11:13]
    s2=s[14:]
    s3=s1+s2
    return int(s3)
data['newCol']=data.Timestamp.apply(f)
del data["Timestamp"]
```

```python
data
```

```python
day = pd.DataFrame(columns=['Latitude','Longitude','newCol'])
day = day.fillna(0) # with 0s rather than NaNs
day
```

```python
night = pd.DataFrame(columns=['Latitude','Longitude','newCol'])
night = night.fillna(0) 
night
```

```python
day=data.loc[(data['newCol'] >= 600) & (data['newCol'] <= 2000)] 
night=data.loc[((data['newCol'] >=0) & (data['newCol'] <= 559)) | ((data['newCol'] >= 2001) & (data['newCol'] < 2400))]
```

```python
day=day.drop('newCol', axis=1)
day
```

```python
night=night.drop('newCol', axis=1)
night
```

```python
xs1=day[day.Latitude< 28.412782]
xs2=day[day.Latitude> 28.88]
xs3=day[day.Longitude< 76.838084]
xs4=day[day.Longitude> 77.323460]
frames1 = [xs1, xs2, xs3,xs4];
xsFinal1 = pd.concat(frames1)
```

```python
xs11=night[night.Latitude< 28.412782]
xs22=night[night.Latitude> 28.88]
xs33=night[night.Longitude< 76.838084]
xs44=night[night.Longitude> 77.323460]
frames2 = [xs11, xs22, xs33,xs44];
xsFinal2 = pd.concat(frames2)
```

```python
#xsFinal.to_csv("Xfinal.csv")
```

```python
for i in xsFinal1.index:
    if(i in day.index):
        day.drop(i,inplace=True)
```

```python
for i in xsFinal2.index:
    if(i in night.index):
        night.drop(i,inplace=True)
```

```python
kmeans = km(n_clusters=18).fit(day)
kmeans1 = km(n_clusters=18).fit(night)
```
