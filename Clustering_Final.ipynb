```python
lis=['District', 'Crime', 'Timestamp']
for c in lis:
    data.drop(c, axis=1, inplace=True)
```

```python

xs1=data[data.Latitude< 28.412782]
xs2=data[data.Latitude> 28.88]
xs3=data[data.Longitude< 76.838084]
xs4=data[data.Longitude> 77.323460]
frames = [xs1, xs2, xs3,xs4];
xsFinal = pd.concat(frames)    



```
